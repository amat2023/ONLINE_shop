```python
#Change the Gender to Str :
```

```python

Online_Shop['Gender_String'] = Online_Shop['Gender'].replace({0: 'Male', 1: 'Female'})
Online_Shop['Gender_String']
```

```python
#Change Time to minutes :
```

```python
Online_Shop['Time_Spent_minutes'] = (Online_Shop['Time_Spent'] / 60).round(2)
print('--------------------------------------------------------------------------------')
```

```python
#Average time by gender:
```

```python
Average_time_by_gender = Online_Shop.groupby('Gender_String')['Time_Spent_minutes'].mean()  	
Average_time_by_gender
```

```python
import matplotlib.pyplot as plt

Average_time_by_gender = Online_Shop.groupby('Gender_String')['Time_Spent_minutes'].mean()  	

Average_time_by_gender.plot(kind='bar', color=['r', 'green'])
plt.title('Average time in minutes by gender')
plt.xlabel('gender')
plt.ylabel('Average time in minutes')
plt.show()
```

```python
#Average Revenue by gender:
```

```python
import matplotlib.pyplot as plt

Average_Revenue_by_gender = Online_Shop.groupby('Gender_String')['Revenue_Total'].mean()  	

Average_Revenue_by_gender.plot(kind='bar', color=['r', 'green'])
plt.title('Average Revenue by gender')
plt.xlabel('gender')
plt.ylabel('Average Revenue')
plt.show()
print('--------------------------------------------------------------------------------')
```

```python
# Total Purchase by Age :
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=Online_Shop, x='Age', y='Purchase_VALUE', marker='o',ms = 5, mec = 'b', mfc = 'r')

plt.title('Purchase value by Age')
plt.xlabel('Age')
plt.ylabel('Purchase VALUE')

plt.show()
```

```python
tuple_temp= []
def Age(Age):
    

    if Age < 18:
        return 'kids'
    elif 18 <= Age < 25:
        return 'Teeneger'
    elif 25 <= Age < 35:
        return 'young adult'
    elif 35 <= Age < 50:
        return 'adults'
    elif 50 <= Age < 65:
        return 'seniors'
   
      
```

```python
Online_Shop ['Age'] = Online_Shop.apply (lambda x: Age (x['Age']), axis=1)
Online_Shop ["Age"]
```

```python
Online_Shop
```

```python
total_purchase_by_age = Online_Shop.groupby('Age')['Purchase_VALUE'].count()
total_purchase_by_age
```

```python
plt.figure(figsize=(10, 8))
total_purchase_by_age.sort_values().plot(kind='bar', color='blue')
plt.title('Total Purchase by Age')
plt.xlabel('Age Group')
plt.ylabel('Total Purchase')
plt.grid(True)
plt.show()
```
